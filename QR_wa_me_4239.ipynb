```python
from textwrap import fill
import qrcode

#WhatsApp number in international format
phone_number ="254742393638" #Kenyan number

#Optional
message = "Hello!"

#WhatsApp link
WhatsApp_link = f"https://wa.me/{phone_number}"

#Create QR code
qr = qrcode.QRCode(
    version=1,
    box_size=10,
    border=5,
error_correction=qrcode.constants.ERROR_CORRECT_H
)
qr.add_data(WhatsApp_link)
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="red")

#Save  image
img.save("WhatApp_QR_Code.png")

#Display image
display(img)
```
